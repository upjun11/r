```python
import matplotlib.pyplot as plt
import numpy as np
```

### 직선 그래프 (Line plot) 를 그려보세요


```python
# plt.plot(x, y) 함수를 사용하여 다음과 같은 그래프를 그려본다
# x, y 축을 갖는 그래프
# x 축 0, 1,2,3,...9
# y 축 x ** 2

# x 축 값 생성 (0부터 9까지)
x = list(range(10))

# y 축 값 생성 (x의 제곱)
y = [i ** 2 for i in x]

# 그래프 그리기
plt.plot(x, y)

# x 축과 y 축에 레이블 추가
plt.xlabel('x')
plt.ylabel('y')

# 그래프 표시
plt.show()
```

```python
# plt.plot(x, y, style) 함수를 사용하여 다음과 같은 그래프를 그려보세요
# x, y 축을 갖고, 빨간색 원('ro')으로 표시되는 그래프
# x 축 0, 1,2,3,...9
# y 축 x ** 2


# x 축 값 생성 (0부터 9까지)
x = list(range(10))

# y 축 값 생성 (x의 제곱)
y = [i ** 2 for i in x]

# 빨간색 원으로 그래프 그리기
plt.plot(x, y, 'g^')

# x 축과 y 축에 레이블 추가
plt.xlabel('x')
plt.ylabel('y')

# 그래프 표시
plt.show()
```

#### 여러 그래프를 겹쳐 그려보세요




```python
# 여러 그래프 겹쳐 그리기
# x 축 0, 1,2,3,...9
# y 축 x*2, x**2, x**3

# x 축 값 생성 (0부터 9까지)
x = list(range(10))

# 각 그래프에 대한 y 축 값 생성 (x의 제곱, x의 세제곱, x의 네제곱)
y_squared = [i ** 2 for i in x]
y_cubed = [i ** 3 for i in x]
y_quad = [i ** 4 for i in x]

# 그래프 그리기
plt.plot(x, y_squared, label='x^2')  # x의 제곱 그래프
plt.plot(x, y_cubed, label='x^3')    # x의 세제곱 그래프
plt.plot(x, y_quad, label='x^4')      # x의 네제곱 그래프

# x 축과 y 축에 레이블 추가
plt.xlabel('x')
plt.ylabel('y')


# 범례 추가
plt.legend()

# 그래프 표시
plt.show()

```
